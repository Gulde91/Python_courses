# Plotting with seaborn

```python
# importing libraries
import seaborn as sns
sns.set()
sns.set(style = "darkgrid")

import numpy as mp
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings 
warnings.filterwarnings("ignore")
```

##### Importing data

```python
data = pd.read_csv("bigmart_data.csv")
data = data.dropna(how = "any")
data["Visibility_Scaled"] = data["Item_Visibility"] * 100
data.head()
```

#### Creating basic plots

```python
sns.lineplot(x = "Item_Weight", y = "Item_MRP", data = data[:50]);
```

##### Bar chart

```python
sns.barplot(x = "Item_Type", y = "Item_MRP", data = data[:5]);
```

##### histogram

```python
sns.distplot(data["Item_MRP"]);
```

##### Boxplot

```python
sns.boxplot(data["Item_Outlet_Sales"], orient = "vertical");
```

##### Violin plot

```python
sns.violinplot(data["Item_Outlet_Sales"], orient = "vertical", color = "magenta");
```

##### Scatter plot

```python
sns.relplot(x = "Item_MRP", y = "Item_Outlet_Sales", data = data[:200], kind = "scatter");
```

##### Hue semantic

```python
sns.relplot(x = "Item_MRP", y = "Item_Outlet_Sales", data = data[:200], hue = "Item_Type");
```

```python
# different line plot for different categories
sns.lineplot(x = "Item_Weight", y = "Item_MRP", hue = "Outlet_Size", data = data[:150]);
```

##### Bubble plot

```python
sns.relplot(x = "Item_MRP", y = "Item_Outlet_Sales", data = data[:200], kind = "scatter", size = "Visibility_Scaled",
           hue = "Visibility_Scaled");
```

##### Category wise sub plot

```python
sns.relplot(x = "Item_Weight", y = "Item_Visibility", hue = "Outlet_Size", style = "Outlet_Size", 
            col = "Outlet_Size", data = data[:100]); 
```

# 2. Advanced categorical plots in seaborn

### Categorical Scatterplots

```python
# strip plot
sns.catplot(x = "Outlet_Size", y = "Item_Outlet_Sales", kind = "strip", data = data[:250]);
```

```python
# swarm plot
sns.catplot(x = "Outlet_Size", y = "Item_Outlet_Sales", kind = "swarm", data = data[:250]);
```

### Categorical Distribution plots

```python
# Box plot
sns.catplot(x = "Outlet_Size", y = "Item_Outlet_Sales", kind = "box", data = data);
```

```python
# Violin plot
sns.catplot(x = "Outlet_Size", y = "Item_Outlet_Sales", kind = "violin", data = data);
```

```python
# Bar plot
sns.catplot(x = "Outlet_Size", y = "Item_Outlet_Sales", kind = "bar", data = data);
```

### Density plots

```python
# distribution of Item visibility
plt.figure(figsize = (7, 7))
sns.kdeplot(data["Item_Visibility"], shade = True);
```

```python
# distribution of Item MRP
plt.figure(figsize = (7, 7))
sns.kdeplot(data["Item_MRP"], shade = False);
```

#### Histogram and Density plot

```python
plt.figure(figsize = (7, 7))
sns.distplot(data["Item_Outlet_Sales"]);
```
