## Visualization with matplotlib

```python
# importing required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# display plots in the notebook itself
%matplotlib inline
```

###### Basics

```python
heigth = [150, 160, 165, 185]
weight = [70, 80, 90, 100]

# draw the plot
plt.plot(heigth, weight)
```

```python
# draw the plot
plt.plot(heigth, weight)
# title
plt.title("Relationship between height and weight")
# x-lab
plt.xlabel("Height")
# y-label
plt.ylabel("weight")
```

```python
calories_burnt = [65, 75, 95, 99]

# draw plot for calories
plt.plot(calories_burnt)

# draw plot for weight
plt.plot(weight)

# adding legends
plt.legend(["Calories Burnt", "Weight"], loc = "lower right")

# adding ticks
plt.xticks(ticks = [0,1,2,3], labels = ['p1', 'p2', 'p3', 'p4']); 
```

```python
# changing figure size
plt.figure(figsize = (15,5))

# plots
plt.plot(calories_burnt)
plt.plot(weight)

plt.legend(["Calories Burnt", "Weight"], loc = "lower right")
plt.xticks(ticks = [0,1,2,3], labels = ['p1', 'p2', 'p3', 'p4']); 
```

```python
# changing line types
plt.plot(calories_burnt, "go") # green color - o shape
plt.plot(weight, "y--")

plt.legend(["Calories Burnt", "Weight"], loc = "lower right")
plt.xticks(ticks = [0,1,2,3], labels = ['p1', 'p2', 'p3', 'p4']); 
```

```python
# Subplots

# Creates 2x2 plots
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (6,6), sharey = True)

# plot on 0 row and 0 column
ax[0].plot(calories_burnt, "go")

# plot on 0 row and 1 column
ax[1].plot(weight)

# sets titles
ax[0].set_title("Calories Burnt")
ax[1].set_title("Weight")

# set ticks
ax[0].set_xticks(ticks=[0,1,2,3]);
ax[1].set_xticks(ticks=[0,1,2,3]);

# set labels
ax[0].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4']);
ax[1].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4']);
```
