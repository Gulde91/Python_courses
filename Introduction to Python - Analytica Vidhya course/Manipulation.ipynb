```python
# importing pandas
import pandas as pd
```

###### Sorting a data frame

```python
# retrieving data
data = pd.read_csv("bigmart_data.csv")

# dropping null values
data = data.dropna(how = "any")

# view data
data.head()
```

```python
# sorting a dataframe
sorted_data = data.sort_values(by = "Outlet_Establishment_Year")

sorted_data.head()
```

```python
data.sort_values(by = "Outlet_Establishment_Year", ascending = False, inplace = True)
data[:5]
```

```python
# sorting by multiple columns
data.sort_values(by = ["Outlet_Establishment_Year", "Outlet_Size"], ascending=False)[:5]
```

```python
# sorting by index
data.sort_index(inplace = True)
data.head()
```

###### Merging data frames

```python
df1 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                  index = [0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                  index = [4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                  index = [8, 9, 10, 11])
```

```python
# concatenating/combining data frames
pd.concat([df1, df2, df3])
```

```python
# combining data frames with keys
df_combined = pd.concat([df1, df2, df3], keys=['x', 'y', 'z'])
df_combined
```

```python
# get second data frame
df_combined.loc['y']
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['C2', 'C3', 'C6', 'C7']},
                  index = [2, 3, 6, 7])
```

```python
# outer join on index
result = pd.concat([df1, df4], axis=1, sort=False)
result
```

```python
# inner join p√• index
result = pd.concat([df1, df4], axis = 1, join = 'inner')
result
```

```python
df_a = pd.DataFrame({'subject_id': ['1', '2', '3', '4', '5'],
                     'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
                     'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']})

df_b = pd.DataFrame({'subject_id': ['4', '5', '6', '7', '8'],
                     'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
                     'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']})

df_c = pd.DataFrame({'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
                     'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]})
```

```python
# merger a og c
pd.merge(df_a, df_c, on = 'subject_id')
```

```python
# outer join
pd.merge(df_a, df_b, on = 'subject_id', how = 'outer')
```

```python
# inner join
pd.merge(df_a, df_b, on = 'subject_id', how = 'inner')
```

```python
# right join
pd.merge(df_a, df_b, on = 'subject_id', how = 'right')
```

```python
# left join
pd.merge(df_a, df_b, on = 'subject_id', how = 'left')
```

###### Applying function

```python
# accessing row wise
data.apply(lambda x: x)
```

```python
# accessing first row
data.apply(lambda x: x[0]) # axis = 0 by default
```

```python
# accessing first column by index
data.apply(lambda x: x[0], axis = 1)
```

```python
# acces by column name
data.apply(lambda x: x["Item_Fat_Content"], axis = 1)
```

```python
# before clipping 
data["Item_MRP"][:5]
```

```python
# clip price if greater than 200
def clip_price(price):
    if price > 200:
        price = 200
    return(price)

# after clipping
data["Item_MRP"].apply(lambda x: clip_price(x))[:5]
```

```python
# before label encoding
data["Outlet_Location_Type"][:5]
```

```python
# label encode city type
def label_encode(city):
    if city == "Tier 1":
        label = 0
    elif city == "Tier 2":
        label = 1
    else:
        label = 2
    return(label)
        
# operate label_encode on every row of Outlet_Location_type
data["Outlet_Location_Type"].apply(lambda x: label_encode(x))[:5]
```

```python
# more sparse syntax - but same result
data["Outlet_Location_Type"].apply(label_encode)[:5]
```

##### Aggregating data

```python
import numpy as np
```

```python
# group price based on item type
price_by_item = data.groupby("Item_Type")

# display first few rows
price_by_item.first()
```

```python
# mean price by item
price_by_item.Item_MRP.mean()
```

```python
# group on multiple columns
multiple_group = data[:10].groupby(["Item_Type", "Item_Fat_Content"])
multiple_group.first()
```

```python
# generate crosstab of Outlet_Size and Outlet_location_type
pd.crosstab(data["Outlet_Size"], data["Outlet_Location_Type"], margins = True)
```

```python
# create pivot table
pd.pivot_table(data, index = ["Outlet_Establishment_Year"], values = "Item_Outlet_Sales")
```

```python
pd.pivot_table(data, index = ["Outlet_Establishment_Year", "Outlet_Location_Type", "Outlet_Size"], values="Item_Outlet_Sales",
              aggfunc=[np.mean, np.median, min, max, np.std])
```
